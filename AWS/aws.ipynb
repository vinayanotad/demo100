```python
#Import pandas packages
import pandas as pd
c=avg=0

data = {'Name' : ['A','B','C','D','E','F'],
        'Age' : [23 , 34 , 45, 23 , 56, 76],
        'Marks': [23, 'NaN', 45, 'NaN' , 34, 23],
        'Gender':['Male','Female','Male','Female','Male','Female']
        }

#Convert into DataFrame
df=pd.DataFrame(data)

#Display data
print(df)

for ele in df['Marks'] :
    if str(ele).isnumeric():
        c+=1
        avg+=ele
        
avg/=c

df=df.replace(to_replace='NaN',value=avg)    #changed values is stored in df

df
```

```python
df['Gender']=df['Gender'].map({'M':0,'F':1}).astype(float)
df
```
