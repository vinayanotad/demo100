```python
filename = input("Enter the filename: ")
char = input("Enter the character to count: ")

with open(filename) as f:
    count = 0
    for line in f:
        for a in line:
            if a == char:
                count += 1

print("The character '{0}' occurs {1} times in the file.".format(char, count))
```

```python
import csv

# define the data
data = [
    ['name', 'class', 'section'],
    ['rakesh', '10', 'D102'],
    ['swarnima', '12', 'A103'],
    ['Samriddhi', '5', 'B104']
]

# open the file in write mode
with open('student.csv', 'w', newline='') as f:
    # create a CSV writer object
    writer = csv.writer(f)

    # write the data to the file
    for row in data:
        writer.writerow(row)

print("File created successfully!")
```

```python
def enroll_student():
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    roll_no = input("Enter student roll number: ")
    email = input("Enter student email: ")
    phone = input("Enter student phone number: ")
    
    # Write student details to file
    with open("students.csv", "a") as file:
        file.write(f"{name}, {age}, {roll_no}, {email}, {phone}\n")
        
    print("Student enrolled successfully!")


def retrieve_students():
    # Read student details from file
    with open("students.txt", "r") as file:
        students = file.readlines()
        
    # Print student details
    if students:
        print("Student details:")
        for student in students:
            print(student.strip())
    else:
        print("No student details found.")


while True:
    print("\n1. Enroll a student\n2. Retrieve all students\n3. Quit")
    choice = input("Enter your choice: ")
    
    if choice == "1":
        enroll_student()
    elif choice == "2":
        retrieve_students()
    elif choice == '3' :
        retrieve_students.()   
    elif choice == "4":
        break
    else:
        print("Invalid choice. Please try again.")
```
